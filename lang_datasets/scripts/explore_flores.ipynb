```python
%load_ext autoreload
%autoreload 2

import os
import random

from constants import LANGUAGES, SELECTED_SENTENCES

SEED = 117
```

```python
flores_dataset_path = "../flores200_dataset/devtest"
file_names = os.listdir(flores_dataset_path) 
languages = [name.split(".")[0] for name in file_names if name.endswith(".devtest")]

# Output the languages into a .txt file if it doesn't already exist
if not os.path.exists("../flores_languages.txt"):
    with open("../flores_languages.txt", "w") as f:
        for lang in languages:
            f.write(f"{lang}\n")
```

```python
print(SELECTED_SENTENCES)
```

```python
# Find the indexes of the selected sentences in the dataset
selected_indices = []
for sentence in SELECTED_SENTENCES:
    found = False
    file_path = os.path.join(flores_dataset_path, f"eng_Latn.devtest")
    with open(file_path, "r", encoding="utf-8") as f:
        lines = f.readlines()
        for i, line in enumerate(lines):
            if line.strip() == sentence:
                selected_indices.append((i))
                found = True
                break
    if not found:
        print(f"Sentence not found: {sentence}")
        continue

assert len(selected_indices) == len(SELECTED_SENTENCES), "Not all sentences were found in the dataset."
print(len(selected_indices), "sentences found in the dataset.")
```

```python
def select_from_other_lang(lang_code, selected_indices):
    lang_path = f"../flores200_dataset/devtest/{lang_code}.devtest"
    with open(lang_path, "r") as f:
        lines = f.readlines()
        selected_sentences = [lines[i].strip() for i in selected_indices]

    # Save the selected sentences to a file
    output_path = f"../flores200_dataset/sample_original/{lang_code}_sample.txt"
    with open(output_path, "w") as f:
        for sentence in selected_sentences:
            f.write(sentence + "\n")
            
    return selected_sentences
```

```python
for lang in LANGUAGES:
    sentences = select_from_other_lang(lang, selected_indices)
    print(f"Selected sentences from {lang}:")
    for sentence in sentences:
        print(sentence)
    assert len(sentences) == len(SELECTED_SENTENCES), f"Not all sentences were found in {lang} dataset."
    print("\n")
```
